```python
N = 5
N_list = [4, 1, 5, 2, 3]
M = 5
M_list = [1, 3, 7, 9, 5]
```

```python
# 아래와 같이 풀면 출력결과는 맞지만 시간복잡도에 의해 시간초과가 나옵니다
for item in M_list:
    if item in N_list:
        print(1)
    else:
        print(0)
```

```python
N_list.sort()

def binary_search(value, start, end):
    if start > end:
        return False

    median = (start + end) // 2
    if N_list[median] > value:
        return binary_search(value, start, median - 1)
    elif N_list[median] < value:
        return binary_search(value, median + 1 , end)
    else:
        return True

for item in M_list:
    if binary_search(item, 0, N - 1):
        print(1)
    else:
        print(0)
```

```python
# 제출 코드
N = int(input())
N_list = list(map(int, input().split()))
M = int(input())
M_list = list(map(int, input().split()))

N_list.sort()

def binary_search(value, start, end):
    if start > end:
        return False

    median = (start + end) // 2
    if N_list[median] > value:
        return binary_search(value, start, median - 1)
    elif N_list[median] < value:
        return binary_search(value, median + 1 , end)
    else:
        return True

for item in M_list:
    if binary_search(item, 0, N - 1):
        print(1)
    else:
        print(0)
```
