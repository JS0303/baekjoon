```python
n,m = map(int, input().split())
castle = []

for _ in range(n):
    castle.append(input())

row = 0
col = 0

for i in range(n):
    if "X" not in castle[i]: # 행별로 경비가 없으며 카운트
        row += 1
for i in range(m):
    col_list = [] # 각 열의 요소를 담을 리스트는 따로 생성
    for j in range(n):
        col_list.append(castle[j][i]) 
    if "X" not in col_list: # 열 단위로 확인하여 경비가 없으면 카운트
        col += 1
# 행과 열 단위로 센 카운트 중 큰 값만큼 경비를 배치
print(max(row, col))
```
