```python
n = int(input())
trop = []

for _ in range(n):
    trop.append(int(input()))


# 왼쪽에서 볼 때
count = 1
max_ = trop[0]

for i in range(0, n-1):
    if trop[i+1] > max_:
        count += 1
        max_ = trop[i+1]
print(count)

# 오른쪽에서 볼 때
count = 1
max_ = trop[n-1]

for i in range(n-1, -1, -1):
    if trop[i] > max_:
        count += 1
        max_ = trop[i]
print(count)
```
